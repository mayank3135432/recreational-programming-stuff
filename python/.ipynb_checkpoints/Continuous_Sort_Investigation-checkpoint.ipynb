```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

```python
listit = lambda f,n : list(map(f,np.arange(0,1,1/n)))
```

```python
# X1=listit(lambda x:1-(2*x-1)**2, 1000)
num=10000
for coeff in range(1,104,3):
    
    X1=listit(lambda x:np.sin(coeff*x), num)
    
    X2=X1.copy()
    X2.sort()
    
    
    # plt.plot(np.arange(0,1,1/num), X1, label="original function")
    plt.plot(np.arange(0,1,1/num), X2, label=f"{coeff}")
    plt.legend()
    # plt.show()
```

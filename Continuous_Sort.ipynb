```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

```python
listit = lambda f,n : list(map(f,np.arange(0,1,1/n)))
```

```python
X1=listit(lambda x:1-(2*x-1)**2, 1000)
X2=X1.copy()
X2.sort()
```

```python
plt.plot(np.arange(0,1,1/1000), X1, label="original function")
plt.plot(np.arange(0,1,1/1000), X2, label="sorted function")
plt.legend()
```
